GoogleMapAPIを格納した'token'というファイルを作成し、アップロードする

```python
with open('token') as f:
    key = f.read()
```

```python
import googlemaps
from datetime import datetime

gmaps = googlemaps.Client(key=key)
```

```python
# find place
find_place_result = gmaps.find_place('立教大学', 'textquery', language='ja')
find_place_result
```

```python
find_place_result['candidates'][0]['place_id']
```

## 場所の情報

```python
# place
place_result = gmaps.place(find_place_result['candidates'][0]['place_id'], language='ja')
place_result
```

## 周辺の情報

```python
# places_nearby
places_nearby_result = gmaps.places_nearby({'lat': 35.73011989999999, 'lng': 139.7034185}, radius=100, language='ja')
places_nearby_result
```
